## Python Supports 3 Catagories Of Data Types : 
1. Basic Types : int, float, complex, boolean, string
2. Container Types : List, Tuples, Sets, Dictionary
3. User Defined Types : Class

```python
# int : 

print(77)

print(1e308) # python can handle very large int (till 10^308)

print(1e309) 
```

```python
# Floating-point numbers in Python

# Regular float
print(7.777)  # Outputs: 7.777

# Scientific notation
print(7.7e307)  # Large number: 7.7 × 10^307
print(7.7e-307)  # Very small number: 7.7 × 10^-307

# Maximum float value in Python
import sys
print(sys.float_info.max)  # Shows the maximum float value Python can handle
```

```python
# boolean : 

print(True)
print(False)
```

```python
# complex : 

print(5+4j)
```

```python
# string : 

print('Hello')
print("World")
print("""Humans""")
```

```python
# List : 

print([1,2,3,4,5])
```

```python
# tuple : 

print((1,2,3,4,5))
```

```python
# sets : 

print({1,2,3,4,5})
```

```python
# Dictionary : 
print({"Name": "Alvi", "Age": 25, "City": "Dhaka"})
```
