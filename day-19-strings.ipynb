## Accessing Substrings from a String : 

```python
# indexing : 

s = 'Bangladesh'

s[7]
```

```python
# positive indexing : 

s = 'Bangladesh'

s[7]
```

```python
# negative indexing : 

s = 'Bangladesh'

s[-1]
```

```python
# Slicing : 

s = 'Hello World'

print(s[0:5])
print(s[6:])
print(s[:5])
print(s[0:7:2])
print(s[::3])
print(s[-5:])
print(s[::-1]) # reverse
print(s[-5::])
print(s[-5:][::-1])
```

## Operators in String : 

1. Arithmetic
2. Relational
3. Logical
4. Loops
5. Membership

```python
# Arithmetic : 

print('Hello'+'World')

print('Hello'+'->'+'World')

print('*'*30)
```

```python
# relational : 

print('Hello' == 'World')

print('Hello'=="Hello")

print('Dhaka'>'Sylhet') # lexiography 

print('A'>'a')
```

```python
# logical : 

'''
empty string -> False
any non empty string-> True

'''

print('' and 'World') # output is empty string 

print('' or 'World')

print('Hello' or 'World')

print(not 'World')

print('Hello' and 'World')
```

```python
# loops : 

s ='Bangladesh'

for i in s:
    print(i , end=' ')
```

```python
for i in s[::2]:
    print(i , end=' ')
```

```python
for i in s[::-1]:
    print(i , end=' ')
```

```python
# membership : 

s = 'Hello World'

print('Hello' in s)

print('ell' in s)

print('W' in s)
```

## String Functions : 

### Common Functions : 
1. len
2. max
3. min
4. sorted

```python
s = 'Bangladesh'

print(len(s))

print(max(s))

print(min(s))

print(sorted(s))

print(sorted(s,reverse=True))
```

### Functions that are only applicable to String : 

```python
# capitalize -> will capitalize the firt char , but not parmanent : 

s='bangladesh'
print(s.capitalize())
print(s)
print('its raining outside'.capitalize())
```

```python
# title -> will title case the string : 

print('its raining outside'.title())
```

```python
# upper : 

print('its raining outside'.upper())
```

```python
# swapcase : 

print('Its raiNing Outside'.swapcase())
```

```python
# count : 

print('its raining outside'.count('i'))
print('its raining outside'.count('in'))
```

```python
# find : 

print('its raining outside'.find('r'))
print('its raining outside'.find('raining'))
print('its raining outside'.find('99'))
```

```python
# index : 
print('its raining outside'.index('r'))
print('its raining outside'.index('raining'))
print('its raining outside'.index('99')) # it will throw an error 
```

```python
# endswith / startswith : 

print('its raining outside'.endswith('de'))
print('its raining outside'.endswith('outside'))
print('its raining outside'.startswith('it'))
```

```python
# format : 

print("hello I am {} and I am {} years old".format('Alvi',25))
```

```python
print("hello I am {1} and I am {0} years old".format('Alvi',25))
```

```python
print("hello I am {name} and I am {age} years old".format(name='Alvi',age=25))
```

```python
# isalnum, isalpha , isdecimal , isdigit , isidentifier : 

print("99abc".isalnum()) # is it alpha numeric 

print('a'.isalpha()) # is only alphabetric

print('99'.isdigit())

print('1num'.isidentifier())
```

```python
# Split -> converts the string into a list : 

s='who is Donald Duck ?'

print(s.split())

print(s.split('is'))

print(s.split('x'))
```

```python
# join : 

print(' '.join(['who', 'is', 'Donald', 'Duck', '?']))
print('_'.join(['who', 'is', 'Donald', 'Duck', '?']))
```

```python
# replace : 

print('I am from Dhaka'.replace('Dhaka','Sylhet'))
```

```python
# Strip : 

name = '               Alvi                 '

print(name.strip())
```
