## Literal is a raw data given to a vairable . In Python there are various type of literals: 
1. Numeric
2. String
3. Boolean
4. Special

```python
# 1. Numeric Literals : 

a = 0b1010 # Binary Literals
b = 70 # decimal literal
c = 0o310 # Octal literal
d = 0x12c # hexadecimal literal

print (a,b,c,d)


# Float Literals : 

e = 77.99
f = 77.88e30
g = 77.66e-7

print(e,f,g)

# complex literals : 

x = 7+5j 
print(x ,  x.imag , x.real)
```

```python
# 2. String Literals : 

s= 'Hello'
s1= "Bangladesh"
s2="A"
multiLine ="""Its a multiline String"""
unicode = u"\U0001F606"
raw_str = r"C:\Users\name"  # Output: C:\Users\name (backslashes preserved)

print(s)
print(s1)
print(s2)
print(multiLine)
print(unicode)
print(raw_str)

```

```python
# 3. Boolean Literals : 

print(True + 1)
print(False + 1)
```
