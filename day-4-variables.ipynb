```python
# Python Variables are Dynamically typed : 

# Dynamically typed -> no need to declare variable type before use : a = 7
# Statically typed -> need to declare variable type before use : int a = 7; 

name = ' Alvi'

print(name)
```

```python
# Python variables has Dynamic Binding feature : 

# Dynamic Binding -> one variable can be used to store different types of variable 
# Static Binding -> you cant (int a = 7 , a = 'A'(error))

x = 'A'
x = 77
x = True
```

```python
# special way to declare easily : 

a,b,c=5,6,7

print(a)
print(b)
print(c)
```

```python
a=b=c=1

print(a)
print(b)
print(c)
```
