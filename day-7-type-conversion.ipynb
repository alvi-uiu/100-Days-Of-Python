```python
# Type Conversion means to convert to one data type to another (if possible)

# There can be two types of type conversion : 

# 1. implicit -> where python automatically does the conversion based on the type of data stored
# 2. explicit -> where user explicitly does the conversion 
```

```python
# implicit : 

print(7+9.5)

print(7 + 4+5j)
```

```python
# explicit : 

a= int(7.7)
b=int('88')

c=float(88)

d=str(99.99)

e=bool(1)

f = list('Hello')

# g = int('alvi') not possible -> incompatibele

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```

```python
# type conversion is not parmanent operation  : 

a = 7

print(float(a))

print(a)

```

```python
# do type conversion while taking input if you know what type of data to take : 

a = int(input('enter first num : '))
b = int(input('enter second num : '))

print(a+b)
```
