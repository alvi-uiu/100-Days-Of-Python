```python
# we can print anything by print function : 
print("Hello World ! ")

print(7)

print(7.77)

print(True)
```

```python
# we can also print multiple items at a time : 

print('Hasina',"Ousted",'on',36,'July')
```

```python
# in each function in python, if we take the cursor inside the function and hit : shift+TAB 
# we will see the inner working of that funciton 
# similarly for print function there is a separator (sep=' ') for which after every word a 
# space is printed . But we can modify it : 
print('Hasina',"Ousted",'on',36,'July', sep='-')
```

```python
# if we dont want new line : 

print('Hasina',end=' ')
print('Ousted')
```
