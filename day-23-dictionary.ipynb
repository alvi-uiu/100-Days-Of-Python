##  Dictionary : -> Key : value : 
1. Dictionary has no indexing
2. Dictionary is mutable
3. keys should be immutable , values can be mutable
4. keys should be unique

Mutable -> Lists, Sets , Dictionary  (can be changed after creation)

Immutable -> int,float,string,tuples,boolean,complex (can not be changed after creation)


```python
# Create Dict : 

# empty dict : 

D = {}
print(type(D))

# normal dict : 

D={"Name":"Alvi","Age":25}
print(D)

# key should be immutabele in dict : 

# D={[1,2,3]:'A'} -> error because list is mutable

D={(1,2,3):'A'} # this will work as tuple is immutable

# keys should be unique : 
D={'Name':'Alvi','Name':'Siam'} # new val updated on the same key 
print(D)

# 2d dict : 

D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}
```

```python
# Accessing Items : 

D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}

print(D['Age'])

print(D['Marks']['Physics'])
```

```python
# Editing items : 

D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}

D['Name']='Siam'

print(D)

D['Marks']['Physics']=88

print(D)

# add new key - val pair : 

D['City']='Dhaka'

D['Marks']['Biology']=86

print(D)
```

```python
# Delete : 

D1 ={'name':'Alvi'}
# delete whole dict : 
del D1 

D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}

# delete a specefic item : 
del D['Age']
print(D)

# clear -> makes an empty dict : 
D.clear()
print(D)
```

```python
# Operatons : 

# ARITHMATIC , MULTIPLICATION WONT WORK ON DICT 

# loops : 
D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}

# this will print only keys : 
for i in D : 
    print(i)

# this will print both keys and values : 

for i in D : 
    print(i,"->",D[i])

# membership -> only works on keys : 

print('Name' in D)
print('Age' in D)
```

```python
# Functions : 

D={'Name':'Alvi','Age':25, 'Marks':{'Math':100,'Physics':99}}

print(len(D))
print(min(D))
print(max(D))
print(sorted(D))
```

```python
# function only for dict : 

print(D.keys())

print(D.values())
```
