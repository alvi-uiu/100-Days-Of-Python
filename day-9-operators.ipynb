## Operators are used to perform operations on variables and values . Python has following operators : 
1. Arithmetic
2. Comparison
3. Logical
4. Bitwise
5. Assignment
6. Identity
7. Membership

### 1. Arithmetic Operators : 

```python
x = 5
y = 2

print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x//y)
print(x**y)
print(x%y)
```

### 2. Comparison : 

```python
print(x==y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
print(x!=y)
```

### 3. Logical : 

```python
x = True
y = False 

print(x or y)
print(x and y)
print(not x)
```

### 3. Bitwise : 

```python
x = 2
y = 3 

print(x & y)
print(x | y)
print(x >> y)
print(x << y)
print(x or y)
print(~y)
```

### 4. Assignment : 

```python
a = 77
a += 3
a /= 5
a *= 5
a %= 3

print(a)
```

### 5. Identity : 

```python
# identity operator (is) checks if the two variables are in same memory location or not : 

a = 9
b = 9 
print(a is b) # in some cases python smartly store same location if the values are same

c=[1,2,3]
d=[1,3,4]
print(c is b) # some cases not
```

### 6. Membership : 

```python
# membership operator checks if something is in something or not : 

s = 'Dhaka'
print ('D' in s)
t = (13,45,88)
print(45 in t)
st={55,99,33}
print(100 not in st)
```
